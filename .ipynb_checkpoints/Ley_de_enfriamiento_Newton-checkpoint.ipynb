# Ecuación de la ley de enfriamiento de Newton

Esta ecuación se obtiene a partir del desarrollo de la ecuación diferencial:

$\frac{dT}{dt}$ = $k(T - T_m)$

Usando el método de variables separables:

1. Separamos diferenciales y los térmimos

$\frac{dT}{dt}$ = $k(T - T_m)$

2. Integramos a ambos lados de la ecuación

$\int{\frac{dT}{T - T_m}}$ = $\int{k dt}$

3. Resultado de la integral

$\ln{(T - T_m)} + C$ = $kt + C$

4. Aplicamos exponencial a ambos lados de la ecuación

$e^{\ln{(T - T_m)}}$ = $e^{kt + C}$

5. Resultado del resultado del exponencial

$T - T_m$ = $e^{kt}  e^C$

6. Obtenemos como resultado la siguiente expresión:

$T = Tm + Ce^{kt}$
    
Donde:

#### Variables:
* t = tiempo en minutos (variable independiente)
* T = temperatura del objeto en base al tiempo transcurrido (variable dependiente)
    
#### Constantes:
* Tm = temperatura del ambiente
* C = constante de integración:
    * Se calcula usando el tiempo = 0
* k = constante de proporcial de temperatura:
    * Se calcula usando un resultado conocido con un tiempo distinto a 0.


### Importación de librerías

```python
import matplotlib.pyplot as plt
import math
```

### Cálculo de la constante de integración C
    C = (T - Tm) / math.exp(k*t)

```python
# Temperatura del medio
Tm = 24

# Temperatura Inicial
T = 75

# Tiempo inicial
t = 0

# Por lo que k*t = 0
C = (T - Tm) / math.exp(0)
print("Constante C = ", C)
```

### Cálculo de la constante de temperatura k
    k = (math.log((T - Tm )/C))/t

```python
# Calculamos la constante de temperatura usando los siguientes valores
T = 61
t = 3

k = (math.log((T - Tm )/C))/t
print("Constante k = {:.5f}".format(k))
```

### Gráfico de la función

Para un temperatura ambiente de 24 Celsius y una temperatura inicial de 75 Celsius la gráfica es la siguiente:

```python
# Declaramos los valores de los x e y
minutes = list(range(0,71))
Temps = [24 + 51 * math.exp(k*value_t) for value_t in minutes]

# Declaramos la temperatura ambiente
ambient_temp = [24] * len(minutes)

percent = 0.7

plt.figure(figsize=(20*percent,12*percent))
plt.plot(minutes, Temps, linestyle='-', marker='o', color='red', label='Temperatura según minutos')
plt.plot(minutes, ambient_temp, linestyle='--', marker='.', color='green', label='Temperatura Ambiente')
plt.legend(loc='upper right', fontsize=15)
plt.title('Ley de enfriamiento de Newton', fontsize=30)
plt.yticks([20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75])
plt.xlabel('Minutos', fontsize=20 )
plt.ylabel('Temperatura en Celsius', fontsize=20)
plt.show()
```
